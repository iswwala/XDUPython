4. \_\_name\_\_变量是一个内置的变量，表示当前模块名称；  
   脚本直接运行时其值为\_\_main\_\_，作为模块被导入时其值为模块文件名，  
   可以用于判断此时模块是否是作为主程序在运行：  
    ```    
    if __name__ == '__main__':
        #只在作为主程序时运行  
    ```

8. `'There are ' + 12 + 'months in a year'  `   
    12是int类型，不能用+和字符串拼接（TypeError: can only concatenate str (not "int") to str）  
    可以改成  
    `'There are ' + str(12) + 'months in a year' `
